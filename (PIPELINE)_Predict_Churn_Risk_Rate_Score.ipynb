```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("churn.csv")
df
```

```python
df.columns
```

```python
selected_col=['age','gender','region_category','membership_category',
              'joined_through_referral','avg_time_spent','avg_transaction_value',
              'avg_frequency_login_days','points_in_wallet','used_special_discount',
              'offer_application_preference','past_complaint','complaint_status',
              'feedback','churn_risk_score']
df=df[selected_col]
df
```

### Data Engineer

```python
df['joined_through_referral'].unique()
```

```python
def replace_question_mark(value):
    if value == '?':
        return 'NaN'
    return value

df['joined_through_referral'] = df['joined_through_referral'].apply(replace_question_mark)

print(df['joined_through_referral'].unique())
```

```python
df['avg_frequency_login_days']
```

```python
df['avg_frequency_login_days'] = df['avg_frequency_login_days'].apply(lambda x:"NaN" if x == 'Error' else x)
df['avg_frequency_login_days'].describe()
```

```python
# Mengganti nilai "unknown" dalam kolom "gender" dengan modus
mode_gender = df['gender'].mode()[0]
df['gender'] = df['gender'].replace('Unknown', 'NaN')
df['gender'].unique()
```

```python
df["churn_risk_score"].unique()
```

```python
#mengubah nilai churn rate -1 jadi 1
df['churn_risk_score'] = df['churn_risk_score'].apply(lambda x:1 if x == -1 else x)
df['churn_risk_score'].unique()
```

```python
# Membuat semua nilai negatif menjadi positif pada kolom avg_time_spent dan point_in_wallet
df['avg_time_spent'] = df['avg_time_spent'].apply(lambda x: abs(x))
df['points_in_wallet'] = df['points_in_wallet'].apply(lambda x: abs(x))
```

## **SPLITTING DATA **

```python
X = df.drop(columns = ["churn_risk_score"])
y = df.churn_risk_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y, random_state= 0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.columns
```

```python
X_train.info()
```

## Preprocessor

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler, FunctionTransformer, OneHotEncoder
```

```python
numerical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy= "mean")),
    ("scaler", StandardScaler())
])


categorical_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy= "most_frequent")),
    ("encode", OneHotEncoder())
])
```

```python
X_train.head()
```

## Preprocessor

```python
from sklearn.compose import ColumnTransformer
```

```python
X_train.columns
```

```python
# Daftar fitur numerik
numerical_features = ["age", "avg_time_spent", "avg_transaction_value", "avg_frequency_login_days", "points_in_wallet"]

# Daftar fitur kategorikal
categorical_features = ["gender", "region_category", "membership_category", "joined_through_referral",
                        "used_special_discount", "offer_application_preference", "past_complaint",
                        "complaint_status", "feedback"]
```

```python
preprocessor = ColumnTransformer([
    ("numeric", numerical_pipeline, numerical_features),
    ("categoric", categorical_pipeline, categorical_features)
])
```

```python
preprocessor
```

## Pipeline

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
RF_pipeline = Pipeline([
    ("prep", preprocessor),
    ("RF", RandomForestClassifier())
])
```

```python
KNN_pipeline = Pipeline([
    ("prep", preprocessor),
    ("KNN", KNeighborsClassifier())
])
```

```python
SVM_pipeline = Pipeline([
    ("prep", preprocessor),
    ("SVM", SVC())
])
```

```python
DT_pipeline = Pipeline([
    ("prep", preprocessor),
    ("DT", DecisionTreeClassifier())
])
```

```python
LR_pipeline = Pipeline([
    ("prep", preprocessor),
    ("LR", LogisticRegression())
])
```

```python
KNN_pipeline
```

```python
RF_pipeline.fit(X_train, y_train)
```

```python
RF_pipeline.score(X_train, y_train), RF_pipeline.score(X_test, y_test)
```

```python
KNN_pipeline.fit(X_train, y_train)
```

```python
KNN_pipeline.score(X_train, y_train), KNN_pipeline.score(X_test, y_test)
```

```python
SVM_pipeline.fit(X_train, y_train)
```

```python
SVM_pipeline.score(X_train, y_train), SVM_pipeline.score(X_test, y_test)
```

```python
DT_pipeline.fit(X_train, y_train)
```

```python
DT_pipeline.score(X_train, y_train), DT_pipeline.score(X_test, y_test)
```

```python
LR_pipeline.fit(X_train, y_train)
```

```python
LR_pipeline.score(X_train, y_train), LR_pipeline.score(X_test, y_test)
```

### Grid SearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
   'SVM__C': np.arange (1.0,11, 1.0),
   'SVM__gamma': ['scale', 'auto', 'float']

}

model = GridSearchCV(SVM_pipeline, parameter, cv = 3, n_jobs= -1, verbose=1)
model.fit(X_train, y_train)
```

```python
pd.DataFrame(model.cv_results_).sort_values("rank_test_score")
```

```python
model.best_params_
```

```python
model.score(X_train, y_train), model.score(X_test, y_test)
```
